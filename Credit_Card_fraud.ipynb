```python
X=df.loc[:,'Time':'Amount']
y=df['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
#from imblearn.over_sampling import SMOTE
#smote=SMOTE()
#X_train,y_train=smote.fit_resample(X_train,y_train)
y_train.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
pred_prob=clf.predict_proba(X_test)
print(classification_report(y_test, y_pred))
```

```python
print(roc_auc_score(y_test,pred_prob[:,1]))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
predicted = model.predict(X_test)
pred_prob2=model.predict_proba(X_test)
print(classification_report(y_test, predicted))
print(roc_auc_score(y_test,pred_prob2[:,1]))
```

```python
from xgboost import XGBClassifier
modelx = XGBClassifier()
modelx.fit(X_train, y_train)
y_pred3 = modelx.predict(X_test)
pred_prob3=modelx.predict_proba(X_test)
print(classification_report(y_test, y_pred3))
print(roc_auc_score(y_test,pred_prob3[:,1]))
```
